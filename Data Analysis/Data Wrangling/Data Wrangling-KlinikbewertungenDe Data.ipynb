<h2 id="data_standardization">3. Data Standardization and Normalization </h2>

'Gesamtzufriedenheit', 'Qualität der Beratung', 'Mediz. Behandlung','Verwaltung und Abläufe', 'Ausstattung und Gestaltung'

Diese sollen normalerweise Number sein.

```python
bewertunglist=['Gesamtzufriedenheit','Qualität der Beratung','Mediz. Behandlung','Verwaltung und Abläufe','Ausstattung und Gestaltung']
sternelist=['Gesamt','Qualität','Behandlung','Verwaltung','Ausstattung']
```

```python
for b in range(len(bewertunglist)):    
    df.loc[df[bewertunglist[b]] =="sehr zufrieden", sternelist[b]] = 100
    df.loc[df[bewertunglist[b]] =="zufrieden", sternelist[b]] = 67
    df.loc[df[bewertunglist[b]] =="weniger zufrieden", sternelist[b]] = 33
    df.loc[df[bewertunglist[b]] =="unzufrieden", sternelist[b]] = 0
```
